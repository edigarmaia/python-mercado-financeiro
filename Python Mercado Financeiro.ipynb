### Dados do Fundamentus - Pegar os dados financeiros das empresas que operam na bolsa

#### Instalando Selenium, pricipal pacote de web-scrapping

```python
!pip install selenium
```

#### Gerenciar driver com webdriver-manager

```python
!pip install webdriver-manager
```

#### Importando os pacotes

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

#### Criando robô para pegar os dados do site Fundamentus

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
url_site = 'https://www.fundamentus.com.br/resultado.php'

driver.get(url_site)
```

#### Pegando os dados da tabela pelo navegador

```python
local_tabela = '/html/body/div[1]/div[2]/table'

elemento = driver.find_element('xpath', local_tabela)
html_tabela = elemento.get_attribute('outerHTML')
html_tabela
```

#### Colocando os dados em uma tabela com o pandas

```python
tabela = pd.read_html(str(html_tabela), thousands = '.', decimal = ',')[0]
tabela
```

#### Fechando o navegador

```python
driver.quit()
```
